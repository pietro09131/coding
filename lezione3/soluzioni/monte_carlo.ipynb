```python
# Import delle librerie necessarie
import numpy as np
import matplotlib.pyplot as plt

# Numero di punti da generare
n_points = 1000

# Generazione di punti casuali uniformemente distribuiti tra -1 e 1
xs = np.random.uniform(-1, 1, n_points)
ys = np.random.uniform(-1, 1, n_points)

# Calcolo della distanza dall'origine
distance_from_origin = np.sqrt(xs**2 + ys**2)

# Determinazione dei punti interni ed esterni al cerchio
is_inside = distance_from_origin < 1
colors = np.where(is_inside, 'orange', 'lightblue')

# Calcolo del valore di π usando il metodo Monte Carlo
n_inside = np.sum(is_inside)
pi_estimate = 4 * n_inside / n_points

# Creazione del grafico
fig, ax = plt.subplots(figsize=(6, 6))
ax.set_title(r"Stima di $\pi$: " + str(pi_estimate), fontsize=14)
ax.set_aspect('equal')

# Disegno del cerchio di riferimento
unit_circle = plt.Circle((0, 0), 1, color='green', fill=False, linewidth=2)
ax.add_artist(unit_circle)

# Visualizzazione dei punti
ax.scatter(xs, ys, c=colors, s=10, alpha=0.8)

# Etichette degli assi e griglia
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
ax.set_xlabel("x", fontsize=12)
ax.set_ylabel("y", fontsize=12)
ax.grid(True, linestyle='--', alpha=0.7)

# Mostra il grafico
plt.tight_layout()
plt.show()
```
